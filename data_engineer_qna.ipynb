**Find factorial of a number**

```python
def fact(n):
  if n == 0:
    return 1
  else:
    return n * fact(n-1)
fact(4)
```

```python
def factorial(n):
    result = 1
    for i in range (1, n+1):
        result *= i
    return result
factorial(4)
```

**Check if a number is prime or not**

```python
def prime(n):
    if n <= 1:
        return False
    for i in range(2, n):
        if n % i == 0:
            return False
    return True
prime(11)
```

**Palindrome Check**

```python
# A string/number is a palindrome if it reads the same backward as forward. For example, "racecar" is a palindrome, but "racecars" is not.

s = "racecar"
if s == s[::-1]:
  print("palindrome")
else:
  print("not palindrome")
```

**Second highest distinct salary overall**

```python
%sql
with cte as (
select * , dense_rank() over(order by salary desc) rn from employee
)
select * from cte where rn = 2
```

**Top 3 earners per derpartment (with emp_name)**

```python
%sql
with cte as (
select * , dense_rank() over(partition by dept order by salary desc) rn from employee
)
select * from cte where rn between 1 and 3
```

**Flag employees above department average salary**

```python
%sql
with cte as (
select emp_id, emp_name, salary, avg(salary) over(partition by dept) dpt_avg_salary,
case when salary > avg(salary) over(partition by dept) then True else False end sal_greater_than_dpt_avg
from employee
)
select * from cte where sal_greater_than_dpt_avg = 'true'
-- select * from employee where sal_greater_than_dpt_avg = True
```

## **Pyspark coding question and answer**

```python
data = [
    (1, "Alice", "TXN1001", 250, "2025-09-20"),
    (1, "Alice", "TXN1001", 250, "2025-09-20"),  # exact duplicate row
    (1, "Alice", "TXN1002", 300, "2025-09-21"),  # same customer, diff txn
    (2, "Bob", "TXN1003", 400, "2025-09-20"),
    (2, "Bob", "TXN1003", 400, "2025-09-20"),    # exact duplicate row
    (3, "Cathy", "TXN1004", 150, "2025-09-21"),
    (4, "Dan", "TXN1005", 200, "2025-09-22"),
    (4, "Dan", "TXN1006", 500, "2025-09-23"),    # same customer, diff txn
]
df = spark.createDataFrame(data, ["customer_id","name", "txn_id", "amount", "txn_date"])
df.show()
```

**Remove duplicate rows**

```python
df.distinct().show()
```

**Exact duplicate records**

```python
from pyspark.sql.functions import *
df.groupBy(df.columns).count().filter(col("count")>1).show()
```

**Get unique customer (by customer_id)**

```python
df.dropDuplicates(['customer_id']).show()
```

**Get unique transaction per customer (customer_id, txn_id)**

```python
df.dropDuplicates(["customer_id", "txn_id"]).show()
```

**Find customer with more than 1 transaction**

```python
df.dropDuplicates(["customer_id", "txn_id"]).groupBy("customer_id").count().filter(col("count")>1).show()
```

**Find duplicate transactions (where txn_id repeats)**

```python
df.groupBy("txn_id").count().filter(col("count")>1).show()
```
